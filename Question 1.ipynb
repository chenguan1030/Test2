```python
# Importing necessary libraries
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

#differential equation
def dydt(y,t):
    return (-2*y*t)

#initial condition
y0 = 1

#create the x-axis that start from 0 until 2 with 100 interval for the integration
start = 0
end = 2
t = np.linspace(start,end,100)

#integration of differential equation
y=sc.integrate.odeint(dydt,y0,t)

#Plotting the result
plt.figure()
plt.plot(t,y,'r',label='ODEs')
plt.title('Test 1 Question 1')
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
